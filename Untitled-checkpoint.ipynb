```python
# Define empty lists to store metrics
mse_list = []
mae_list = []
r2_list = []
```

```python

# Function to calculate various metrics such as accuracy, precision etc
def calculateMetrics(algorithm, predict, testY):
    testY = testY.astype('int')
    predict = predict.astype('int')
    
    mse = mean_squared_error(testY, predict)
    mae = mean_absolute_error(testY, predict)
    r2 = r2_score(testY, predict) * 100
    
    # Append calculated metrics to lists
    mse_list.append(mse)
    mae_list.append(mae)
    r2_list.append(r2)
    
    print(algorithm + ' Mean Squared Error: ' + str(mse))
    print(algorithm + ' Mean Absolute Error: ' + str(mae))
    print(algorithm + ' R^2 Score: ' + str(r2))
    
    # Create a scatter plot
    plt.figure(figsize=(8, 8))
    plt.scatter(predict, testY, color='blue')
    plt.plot([min(testY), max(testY)], [min(testY), max(testY)], linestyle='--', color='red', lw=2)  # Identity line
    # Set labels and title
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Regression Performance')
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import joblib
import os

if os.path.exists('LinearRegression_model.pkl'):
    # Load the trained model from the file
    lr = joblib.load('LinearRegression_model.pkl')
    print("Model loaded successfully.")
    predict = lr.predict(X_test)
    calculateMetrics("Linear Regression", predict, y_test)
else:
    # Train the model (assuming X_train and y_train are defined)
    lr = LinearRegression(fit_intercept=False,   # Do not fit the intercept — this can reduce accuracy if the data isn't centered.
    copy_X=False,          # Avoid copying X — doesn't directly affect accuracy but speeds up training.
    n_jobs=2,           # Single-threaded
    positive=False )
    
    lr.fit(X_train, y_train)
    # Save the trained model to a file
    joblib.dump(lr, 'LinearRegression_model.pkl')
    print("Model saved successfully.")
    predict = lr.predict(X_test)
    calculateMetrics("Linear Regression", predict, y_test)
```

## Proposed Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
import joblib
import os

if os.path.exists('GradientBoosting_model.pkl'):
    # Load the trained model from the file
    gbr = joblib.load('GradientBoosting_model.pkl')
    print("Model loaded successfully.")
    predict = gbr.predict(X_test)
    calculateMetrics("Gradient Boosting Regressor", predict, y_test)
else:
    # Train the model (assuming X_train and y_train are defined)
    gbr = GradientBoostingRegressor()
    gbr.fit(X_train, y_train)
    # Save the trained model to a file
    joblib.dump(gbr, 'GradientBoosting_model.pkl')
    print("Model saved successfully.")
    predict = gbr.predict(X_test)
    calculateMetrics("Gradient Boosting Regressor", predict, y_test)
```
